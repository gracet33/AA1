```python
# Create a new 'MembershipLevel' column based on conditions
df['MembershipLevel'] = "Not Member"  # Default level

# Upgrade to Bronze for customers in city tiers 1 and 2 with tenure > 2 and hours spent on app > 2
df.loc[(df['CityTier'].isin([1, 2])) & (df['Tenure'].gt(2)) & (df['HourSpendOnApp'].gt(2)), 'MembershipLevel'] = 'Bronze'

# Upgrade to Silver for customers in city tiers 2 and 3 with tenure > 4 and hours spent on app > 2
df.loc[(df['CityTier'].isin([2, 3])) & (df['Tenure'].gt(4)) & (df['HourSpendOnApp'].gt(2)), 'MembershipLevel'] = 'Silver'

# Upgrade to Gold for customers in city tiers 2 and 3 with tenure > 8 and hours spent on app > 3
df.loc[(df['CityTier'].isin([2, 3])) & (df['Tenure'].gt(8)) & (df['HourSpendOnApp'].gt(3)), 'MembershipLevel'] = 'Gold'

# Upgrade to Platinum for customers in city tier 3 with the highest tenure and hours spent on app > 3
df.loc[(df['CityTier'].eq(3)) & (df['Tenure'].gt(16)) & (df['HourSpendOnApp'].gt(3)), 'MembershipLevel'] = 'Platinum'

# Print the first few rows of the Da
```
